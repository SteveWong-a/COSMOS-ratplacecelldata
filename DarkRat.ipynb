```python
# -*- coding: utf-8 -*-
"""
Colab script to train two MLPs for ordinal classification of head direction
based on calcium signals, using `dirbin` (1…K) converted to 0…K-1 as labels.
"""

# 1. Mount Google Drive (force remount if already mounted)
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# 2. Imports
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# 3. Utility: Load NPZ and prepare features & labels
def load_dataset(npz_path):
    """
    Loads calcium signals (X) and integer direction‐bin labels (y) from an .npz file.
    Converts `dirbin` from 1…K to 0…K-1.
    """
    data = np.load(npz_path)
    X = data['calcium_signal']                 # shape: (n_samples, n_neurons)
    y = data['dirbin'].astype(int).ravel() - 1 # make labels 0…K-1
    return X, y

# 4. Define MLP for classification
class MLP(nn.Module):
    def __init__(self, input_dim, hidden_dims, num_classes):
        super().__init__()
        layers = []
        prev = input_dim
        for h in hidden_dims:
            layers += [nn.Linear(prev, h), nn.ReLU(inplace=True)]
            prev = h
        layers.append(nn.Linear(prev, num_classes))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

# 5. Custom metric: MSE between predicted and true class indices
def compute_mse(logits, targets):
    pred = torch.argmax(logits, dim=1).float()
    true = targets.float()
    return torch.mean((pred - true) ** 2)

# 6. Training loop
def train_model(X, y, name,
                hidden_dims=[128, 64],
                batch_size=64,
                lr=1e-3,
                epochs=30,
                test_size=0.2,
                random_state=42,
                device=None):
    device = device or ('cuda' if torch.cuda.is_available() else 'cpu')

    # Split and scale
    X_tr, X_val, y_tr, y_val = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )
    scaler = StandardScaler().fit(X_tr)
    X_tr = scaler.transform(X_tr)
    X_val = scaler.transform(X_val)

    # DataLoaders
    train_ds = TensorDataset(torch.from_numpy(X_tr).float(),
                             torch.from_numpy(y_tr).long())
    val_ds   = TensorDataset(torch.from_numpy(X_val).float(),
                             torch.from_numpy(y_val).long())
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_ds, batch_size=batch_size)

    # Model & optim
    input_dim   = X.shape[1]
    num_classes = int(y.max()) + 1
    model       = MLP(input_dim, hidden_dims, num_classes).to(device)
    criterion   = nn.CrossEntropyLoss()
    optimizer   = torch.optim.Adam(model.parameters(), lr=lr)

    print(f"\n=== Training MLP for {name} rats ===")
    for epoch in range(1, epochs + 1):
        # Train
        model.train()
        t_loss = t_corr = t_mse = 0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            logits = model(xb)
            loss   = criterion(logits, yb)
            loss.backward()
            optimizer.step()

            t_loss  += loss.item()*xb.size(0)
            preds    = logits.argmax(dim=1)
            t_corr  += (preds==yb).sum().item()
            t_mse   += compute_mse(logits, yb).item()*xb.size(0)

        # Validate
        model.eval()
        v_loss = v_corr = v_mse = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                logits = model(xb)
                loss   = criterion(logits, yb)

                v_loss += loss.item()*xb.size(0)
                preds   = logits.argmax(dim=1)
                v_corr += (preds==yb).sum().item()
                v_mse  += compute_mse(logits, yb).item()*xb.size(0)

        # Metrics
        n_tr = len(train_loader.dataset)
        n_val = len(val_loader.dataset)
        print(f"Epoch {epoch:02d} | "
              f"Train L={t_loss/n_tr:.4f} Acc={t_corr/n_tr:.4f} MSE={t_mse/n_tr:.4f} || "
              f"Val   L={v_loss/n_val:.4f} Acc={v_corr/n_val:.4f} MSE={v_mse/n_val:.4f}")

    # Collect preds for confusion matrix
    all_p, all_t = [], []
    model.eval()
    with torch.no_grad():
        for xb, yb in val_loader:
            xb = xb.to(device)
            logits = model(xb)
            all_p.append(logits.argmax(dim=1).cpu())
            all_t.append(yb)

    y_pred = torch.cat(all_p).numpy()
    y_true = torch.cat(all_t).numpy()

    print(f"\n*** Final {name} Results ***")
    print(f"Val Accuracy: {np.mean(y_pred==y_true):.4f}\n")
    return model, scaler, y_true, y_pred

# 7. Paths & Load
reverse_path = '/content/drive/MyDrive/COSMOS_010reverse.npz'
perfect_path = '/content/drive/MyDrive/COSMOS_010perfect.npz'

X_rev,  y_rev  = load_dataset(reverse_path)
X_perf, y_perf = load_dataset(perfect_path)

# 8. (Optional) Quick sanity check
for tag, y in [('Reversal', y_rev), ('Perfect', y_perf)]:
    print(f"{tag} bins 0…K-1 → {np.unique(y)}; counts: {np.bincount(y)}")

# 9. Train & evaluate
model_rev, _, y_rev_t,  y_rev_p  = train_model(X_rev,  y_rev,  name='Reversal')
model_perf, _, y_perf_t, y_perf_p = train_model(X_perf, y_perf, name='Perfect')

# 10. Plot confusion matrices
def plot_cm(y_true, y_pred, title, ax):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(cm)
    disp.plot(cmap='Blues', ax=ax, colorbar=False)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
    ax.set_title(title)

fig, (a1, a2) = plt.subplots(1, 2, figsize=(12,5), constrained_layout=True)
plot_cm(y_rev_t,  y_rev_p,  'Reversal Rats', a1)
plot_cm(y_perf_t, y_perf_p, 'Perfect Rats',  a2)
plt.show()
```

```python
# Colab Script: MLP for Ordinal Classification of Rat Head Position (8 Bins)

# 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# 2. Imports
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# 3. Load & Bin Data into 8 Classes

def load_reversal_data(path, num_bins=8):
    """
    Returns:
      calcium: (N, D) signals
      labels:  (N,) integer bins 0…num_bins-1
      epochs:  (N,) epoch codes
    """
    data        = np.load(path)
    calcium     = data['calcium_signal']
    angles_rad  = data['head_orientation'].ravel()
    epochs      = data['epoch'].astype(int).ravel()

    # Convert from radians to [0,360) degrees
    angles_deg  = np.rad2deg(angles_rad)               # –180…180
    angles360   = np.mod(angles_deg, 360)              # 0…360

    # Uniform bins
    bin_size    = 360.0 / num_bins
    labels      = np.floor(angles360 / bin_size).astype(int)
    labels[labels == num_bins] = num_bins - 1          # guard edge-case

    # Sanity-check distribution
    counts = np.bincount(labels, minlength=num_bins)
    print("Bin counts:", counts)

    return calcium, labels, epochs
# usage
rev_path  = '/content/drive/MyDrive/COSMOS_010reverse.npz'
num_bins  = 8
X_all, y_all, epoch_all = load_reversal_data(rev_path, num_bins)

# Split by epochs
train_idx  = np.where(epoch_all == 3)[0]
test1_idx  = np.where(epoch_all == 1)[0]
test3_idx  = np.where(epoch_all == 3)[0]

X_train, y_train = X_all[train_idx], y_all[train_idx]
X_test1, y_test1 = X_all[test1_idx], y_all[test1_idx]
X_test3, y_test3 = X_all[test3_idx], y_all[test3_idx]

# 4. Define MLP & Utilities
class MLP(nn.Module):
    def __init__(self, input_dim, hidden_dims, num_classes):
        super().__init__()
        layers = []
        prev = input_dim
        for h in hidden_dims:
            layers += [nn.Linear(prev, h), nn.ReLU(inplace=True)]
            prev = h
        layers.append(nn.Linear(prev, num_classes))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

def compute_mse(logits, targets):
    preds = logits.argmax(dim=1).float()
    return torch.mean((preds - targets.float())**2)

# 5. Prepare for Training
device      = 'cuda' if torch.cuda.is_available() else 'cpu'
scaler      = StandardScaler().fit(X_train)
X_tr_scaled = scaler.transform(X_train)

train_ds    = TensorDataset(torch.from_numpy(X_tr_scaled).float(),
                            torch.from_numpy(y_train).long())
train_loader= DataLoader(train_ds, batch_size=64, shuffle=True)

input_dim   = X_train.shape[1]
num_classes = num_bins

model     = MLP(input_dim, [128, 64], num_classes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
num_epochs= 50

# 6. Training Loop (Epoch 3 Only)
for epoch in range(1, num_epochs + 1):
    model.train()
    running_loss = running_corr = running_mse = total = 0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        logits = model(xb)
        loss   = criterion(logits, yb)
        loss.backward()
        optimizer.step()

        preds = logits.argmax(dim=1)
        running_loss += loss.item() * xb.size(0)
        running_corr += (preds == yb).sum().item()
        running_mse  += compute_mse(logits, yb).item() * xb.size(0)
        total        += xb.size(0)

    print(f"Epoch {epoch:02d} | "
          f"Loss {(running_loss/total):.4f}  "
          f"Acc {(running_corr/total):.4f}  "
          f"MSE {(running_mse/total):.4f}")

# 7. Evaluation Function
def evaluate(X, y, title):
    Xs    = scaler.transform(X)
    ds    = TensorDataset(torch.from_numpy(Xs).float(),
                          torch.from_numpy(y).long())
    loader= DataLoader(ds, batch_size=64)

    model.eval()
    correct = mse_sum = 0
    preds_all, true_all = [], []

    with torch.no_grad():
        for xb, yb in loader:
            xb, yb = xb.to(device), yb.to(device)
            logits = model(xb)
            preds  = logits.argmax(dim=1)
            correct += (preds == yb).sum().item()
            mse_sum += compute_mse(logits, yb).item() * xb.size(0)
            preds_all.append(preds.cpu())
            true_all.append(yb.cpu())

    total = len(y)
    acc   = correct / total
    mse   = mse_sum / total
    print(f"{title} → Accuracy: {acc:.4f}, MSE: {mse:.4f}")

    return (torch.cat(true_all).numpy(), torch.cat(preds_all).numpy())

y1_true, y1_pred = evaluate(X_test1, y_test1, "Test Epoch 1")
y3_true, y3_pred = evaluate(X_test3, y_test3, "Test Epoch 3")

import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

num_bins   = 8
bin_labels = [f"{i*45}°" for i in range(num_bins)]
full_labels= np.arange(num_bins)   # [0,1,…,7]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True)

for ax, y_t, y_p, title in [
    (ax1, y1_true, y1_pred, 'Epoch 1'),
    (ax2, y3_true, y3_pred, 'Epoch 3'),
]:
    # 1. Explicitly request an 8×8 CM
    cm = confusion_matrix(y_t, y_p, labels=full_labels)

    # 2. Create the display with exactly 8 labels
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=bin_labels)
    disp.plot(ax=ax, cmap='Blues', colorbar=False)

    # 3. Overwrite ticks to be exactly 0–7
    ax.set_xticks(full_labels)
    ax.set_yticks(full_labels)
    ax.set_xticklabels(bin_labels, rotation=45)
    ax.set_yticklabels(bin_labels)

    ax.set_title(title)
    ax.set_xlabel('Predicted Bin')
    ax.set_ylabel('Actual Bin')

plt.show()
data   = np.load(rev_path)
"""
angles = data['head_orientation'].ravel()
print("raw head_ori range:", angles.min(), angles.max())
# if this prints something like –3.14 to +3.14, you’re in radians!

# check how many samples in each new bin
labels = load_reversal_data(rev_path, num_bins)[1]
print("counts per bin:", np.bincount(labels, minlength=num_bins))
"""
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# 1. Load & preprocess
# -----------------------------------------------------------------------------
data        = np.load('/content/drive/MyDrive/COSMOS_010perfect.npz')
X_all       = data['calcium_signal']
angles_rad  = data['head_orientation'].ravel() % (2*np.pi)
tnum        = np.abs(data['tnum'].ravel()).astype(int)

# 2. Radial bins (8 cyclic ordinal classes)
num_bins = 8
edges    = np.linspace(0, 2*np.pi, num_bins+1)
labels   = np.digitize(angles_rad, edges) - 1
labels[labels == num_bins] = num_bins - 1

# 3. Split indices by tnum ranges
idx_25_32  = np.where((tnum >= 25) & (tnum <= 32))[0]
idx_1_8    = np.where((tnum >= 1 ) & (tnum <= 8 ))[0]
idx_17_24  = np.where((tnum >= 17) & (tnum <= 24))[0]

rng       = np.random.default_rng(0)
perm      = rng.permutation(idx_25_32)
n_hold    = int(0.2 * len(perm))
test2_idx = perm[:n_hold]
train_idx = perm[n_hold:]
test1_idx = idx_1_8
test3_idx = idx_17_24

# 4. Build train/test subsets
X_tr, y_tr = X_all[train_idx], labels[train_idx]
X_t1, y_t1 = X_all[test1_idx], labels[test1_idx]
X_t2, y_t2 = X_all[test2_idx], labels[test2_idx]
X_t3, y_t3 = X_all[test3_idx], labels[test3_idx]

# 5. Drop near‐zero‐variance features (apply same mask to all splits)
train_var = X_tr.var(axis=0)
mask      = train_var > 1e-6
X_tr, X_t1, X_t2, X_t3 = (
    X_tr[:, mask],
    X_t1[:, mask],
    X_t2[:, mask],
    X_t3[:, mask]
)

# -----------------------------------------------------------------------------
# 6. Data Preprocessing: z-score → clip → Min–Max scaling
# -----------------------------------------------------------------------------
# 6.1 Compute per-feature mean & std on TRAIN
mean_tr = X_tr.mean(axis=0)
std_tr  = X_tr.std(axis=0)

# 6.2 Z-score normalize all splits
X_tr_z  = (X_tr  - mean_tr) / std_tr
X_t1_z  = (X_t1  - mean_tr) / std_tr
X_t2_z  = (X_t2  - mean_tr) / std_tr
X_t3_z  = (X_t3  - mean_tr) / std_tr

# 6.3 Clip to ±2 standard deviations
X_tr_clip = np.clip(X_tr_z, -2, 2)
X_t1_clip = np.clip(X_t1_z, -2, 2)
X_t2_clip = np.clip(X_t2_z, -2, 2)
X_t3_clip = np.clip(X_t3_z, -2, 2)

# 6.4 Fit MinMaxScaler on clipped TRAIN only
scaler = MinMaxScaler(feature_range=(0,1)).fit(X_tr_clip)

# 6.5 Transform all clipped splits
X_tr = scaler.transform(X_tr_clip)
X_t1 = scaler.transform(X_t1_clip)
X_t2 = scaler.transform(X_t2_clip)
X_t3 = scaler.transform(X_t3_clip)

# 7. Compute class‐frequency weights
counts        = np.bincount(y_tr, minlength=num_bins)
inv_counts    = 1.0 / (counts + 1e-6)
class_weights = inv_counts / inv_counts.sum() * num_bins
weights_t     = torch.tensor(class_weights, dtype=torch.float)

# 8. Prepare cyclic‐distance matrix
dist = torch.zeros(num_bins, num_bins)
for i in range(num_bins):
    for j in range(num_bins):
        diff       = abs(i - j)
        dist[i, j] = min(diff, num_bins - diff)

# 9. DataLoaders
def make_loader(X, y, bs=64, shuffle=False):
    ds = TensorDataset(torch.from_numpy(X).float(),
                       torch.from_numpy(y).long())
    return DataLoader(ds, batch_size=bs, shuffle=shuffle)

train_loader = make_loader(X_tr, y_tr, shuffle=True)
test1_loader = make_loader(X_t1, y_t1)
test2_loader = make_loader(X_t2, y_t2)
test3_loader = make_loader(X_t3, y_t3)

# -----------------------------------------------------------------------------
# 10. Single-Layer Perceptron (SLP) definition
# -----------------------------------------------------------------------------
class SLP(nn.Module):
    def __init__(self, in_dim, num_cls):
        super().__init__()
        self.fc = nn.Linear(in_dim, num_cls)

    def forward(self, x):
        return self.fc(x)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model  = SLP(X_tr.shape[1], num_bins).to(device)

# 11. Cyclic ordinal loss
class CyclicOrdinalLoss(nn.Module):
    def __init__(self, dist_mat, class_weights):
        super().__init__()
        self.register_buffer('dist_mat', dist_mat)
        self.register_buffer('class_w', class_weights)

    def forward(self, logits, targets):
        p           = torch.softmax(logits, dim=1)
        d           = self.dist_mat[targets]
        sample_loss = (p * d).sum(dim=1)
        w           = self.class_w[targets]
        return (sample_loss * w).mean()

criterion = CyclicOrdinalLoss(dist, weights_t.to(device))
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
num_epochs= 50

# -----------------------------------------------------------------------------
# 12. Training
# -----------------------------------------------------------------------------
for epoch in range(1, num_epochs+1):
    model.train()
    total, correct, loss_sum = 0, 0, 0.0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        logits = model(xb)
        loss   = criterion(logits, yb)
        loss.backward()
        optimizer.step()

        preds     = logits.argmax(dim=1)
        correct  += (preds == yb).sum().item()
        loss_sum += loss.item() * yb.size(0)
        total    += yb.size(0)

    acc      = correct / total
    avg_loss = loss_sum / total
    print(f"Epoch {epoch:02d} | Loss: {avg_loss:.4f} | Acc: {acc:.4f}")

# -----------------------------------------------------------------------------
# 13. Evaluation + confusion matrices (3 test sets)
# -----------------------------------------------------------------------------
def evaluate(loader, title):
    model.eval()
    total, correct = 0, 0
    all_t, all_p   = [], []
    with torch.no_grad():
        for xb, yb in loader:
            xb, yb = xb.to(device), yb.to(device)
            preds  = model(xb).argmax(dim=1)
            correct += (preds == yb).sum().item()
            total   += yb.size(0)
            all_t.append(yb.cpu()); all_p.append(preds.cpu())
    acc = correct / total
    print(f"{title} | Acc: {acc:.4f}")
    return torch.cat(all_t).numpy(), torch.cat(all_p).numpy()

y1_t, y1_p = evaluate(test1_loader,    "Test |tnum| 1–8")
y2_t, y2_p = evaluate(test2_loader, "Hold-out |tnum| 25–32")
y3_t, y3_p = evaluate(test3_loader,  "Test |tnum| 17–24")

fig, axes = plt.subplots(1, 3, figsize=(15,4), constrained_layout=True)
labels_str = [str(i) for i in range(num_bins)]
for ax, yt, yp, ttl in zip(
    axes,
    [y1_t, y2_t, y3_t],
    [y1_p, y2_p, y3_p],
    ["|tnum| 1–8", "Hold-out |tnum| 25–32", "|tnum| 17–24"]
):
    cm   = confusion_matrix(yt, yp, labels=range(num_bins))
    disp = ConfusionMatrixDisplay(cm, display_labels=labels_str)
    disp.plot(ax=ax, cmap="Blues", colorbar=False, xticks_rotation=45)
    ax.set_title(ttl)
    ax.set_xlabel("Pred")
    ax.set_ylabel("True")
    ax.tick_params(labelsize=8)

plt.show()
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# 1. Load & preprocess
# -----------------------------------------------------------------------------
data        = np.load('/content/drive/MyDrive/COSMOS_010perfect.npz')
X_all       = data['calcium_signal']
angles_rad  = data['head_orientation'].ravel() % (2*np.pi)
tnum        = np.abs(data['tnum'].ravel()).astype(int)

# -----------------------------------------------------------------------------
# 2. Radial bins (8 cyclic ordinal classes)
# -----------------------------------------------------------------------------
num_bins = 8
edges    = np.linspace(0, 2*np.pi, num_bins+1)
labels   = np.digitize(angles_rad, edges) - 1
labels[labels == num_bins] = num_bins - 1

# -----------------------------------------------------------------------------
# 3. Split indices by tnum ranges
# -----------------------------------------------------------------------------
idx_train  = np.where((tnum >= 20) & (tnum <= 26))[0]
idx_test1  = np.where((tnum >=  0) & (tnum <=  6))[0]
idx_test2  = np.where((tnum >= 13) & (tnum <= 19))[0]
idx_test3  = idx_train  # In-distribution evaluation (same as train range)

# -----------------------------------------------------------------------------
# 4. Build train/test subsets
# -----------------------------------------------------------------------------
X_tr, y_tr = X_all[idx_train], labels[idx_train]
X_t1, y_t1 = X_all[idx_test1], labels[idx_test1]
X_t2, y_t2 = X_all[idx_test2], labels[idx_test2]
X_t3, y_t3 = X_all[idx_test3], labels[idx_test3]

# -----------------------------------------------------------------------------
# 5. Drop near‐zero‐variance features (apply same mask to all splits)
# -----------------------------------------------------------------------------
train_var = X_tr.var(axis=0)
mask      = train_var > 1e-6
X_tr, X_t1, X_t2, X_t3 = (
    X_tr[:, mask],
    X_t1[:, mask],
    X_t2[:, mask],
    X_t3[:, mask]
)

# -----------------------------------------------------------------------------
# 6. Data Preprocessing: z-score → clip → Min–Max scaling
# -----------------------------------------------------------------------------
mean_tr    = X_tr.mean(axis=0)
std_tr     = X_tr.std(axis=0)

X_tr_z     = (X_tr - mean_tr) / std_tr
X_t1_z     = (X_t1 - mean_tr) / std_tr
X_t2_z     = (X_t2 - mean_tr) / std_tr
X_t3_z     = (X_t3 - mean_tr) / std_tr

X_tr_clip  = np.clip(X_tr_z, -2, 2)
X_t1_clip  = np.clip(X_t1_z, -2, 2)
X_t2_clip  = np.clip(X_t2_z, -2, 2)
X_t3_clip  = np.clip(X_t3_z, -2, 2)

scaler     = MinMaxScaler((0,1)).fit(X_tr_clip)
X_tr        = scaler.transform(X_tr_clip)
X_t1        = scaler.transform(X_t1_clip)
X_t2        = scaler.transform(X_t2_clip)
X_t3        = scaler.transform(X_t3_clip)

# -----------------------------------------------------------------------------
# 7. Class‐frequency weights
# -----------------------------------------------------------------------------
counts        = np.bincount(y_tr, minlength=num_bins)
inv_counts    = 1.0 / (counts + 1e-6)
class_weights = inv_counts / inv_counts.sum() * num_bins
weights_t     = torch.tensor(class_weights, dtype=torch.float)

# -----------------------------------------------------------------------------
# 8. Cyclic‐distance matrix
# -----------------------------------------------------------------------------
dist = torch.zeros(num_bins, num_bins)
for i in range(num_bins):
    for j in range(num_bins):
        d         = abs(i - j)
        dist[i,j] = min(d, num_bins - d)

# -----------------------------------------------------------------------------
# 9. DataLoaders
# -----------------------------------------------------------------------------
def make_loader(X, y, bs=64, shuffle=False):
    ds = TensorDataset(torch.from_numpy(X).float(),
                       torch.from_numpy(y).long())
    return DataLoader(ds, batch_size=bs, shuffle=shuffle)

train_loader = make_loader(X_tr, y_tr, shuffle=True)
test1_loader = make_loader(X_t1, y_t1)
test2_loader = make_loader(X_t2, y_t2)
test3_loader = make_loader(X_t3, y_t3)

# -----------------------------------------------------------------------------
# 10. MLP + Cyclic Ordinal Loss
# -----------------------------------------------------------------------------
class MLP(nn.Module):
    def __init__(self, in_dim, hids, num_cls):
        super().__init__()
        layers = [nn.BatchNorm1d(in_dim)]
        prev   = in_dim
        for h in hids:
            layers += [nn.Linear(prev, h),
                       nn.BatchNorm1d(h),
                       nn.ReLU(inplace=True)]
            prev = h
        layers.append(nn.Linear(prev, num_cls))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

class CyclicOrdinalLoss(nn.Module):
    def __init__(self, dist_mat, class_weights):
        super().__init__()
        self.register_buffer('dist_mat', dist_mat)
        self.register_buffer('class_w', class_weights)

    def forward(self, logits, targets):
        p           = torch.softmax(logits, dim=1)
        d           = self.dist_mat[targets]
        sample_loss = (p * d).sum(dim=1)
        w           = self.class_w[targets]
        return (sample_loss * w).mean()

device    = 'cuda' if torch.cuda.is_available() else 'cpu'
model     = MLP(X_tr.shape[1], [128,64], num_bins).to(device)
criterion = CyclicOrdinalLoss(dist, weights_t.to(device))
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
num_epochs= 50

# -----------------------------------------------------------------------------
# 11. Training loop
# -----------------------------------------------------------------------------
for epoch in range(1, num_epochs+1):
    model.train()
    total, correct, loss_sum = 0, 0, 0.0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        logits = model(xb)
        loss   = criterion(logits, yb)
        loss.backward()
        optimizer.step()

        preds     = logits.argmax(dim=1)
        correct  += (preds == yb).sum().item()
        loss_sum += loss.item() * yb.size(0)
        total    += yb.size(0)

    print(f"Epoch {epoch:02d} | Loss: {loss_sum/total:.4f} | Acc: {correct/total:.4f}")

# -----------------------------------------------------------------------------
# 12. Evaluation + Confusion Matrices
# -----------------------------------------------------------------------------
def evaluate(loader, title):
    model.eval()
    total, correct = 0, 0
    all_t, all_p   = [], []
    with torch.no_grad():
        for xb, yb in loader:
            xb, yb = xb.to(device), yb.to(device)
            preds  = model(xb).argmax(dim=1)
            correct += (preds == yb).sum().item()
            total   += yb.size(0)
            all_t.append(yb.cpu()); all_p.append(preds.cpu())
    print(f"{title} | Acc: {correct/total:.4f}")
    return torch.cat(all_t).numpy(), torch.cat(all_p).numpy()

y1_t, y1_p = evaluate(test1_loader, "|tnum| 0–6")
y2_t, y2_p = evaluate(test2_loader, "|tnum| 13–19")
y3_t, y3_p = evaluate(test3_loader, "|tnum| 20–26")

fig, axes = plt.subplots(1, 3, figsize=(15,4), constrained_layout=True)
labels_str = [str(i) for i in range(num_bins)]
for ax, yt, yp, ttl in zip(
    axes,
    [y1_t, y2_t, y3_t],
    [y1_p, y2_p, y3_p],
    ["|tnum| 0–6", "|tnum| 13–19", "|tnum| 20–26"]
):
    cm   = confusion_matrix(yt, yp, labels=range(num_bins))
    disp = ConfusionMatrixDisplay(cm, display_labels=labels_str)
    disp.plot(ax=ax, cmap="Blues", colorbar=False, xticks_rotation=45)
    ax.set_title(ttl)
    ax.set_xlabel("Pred")
    ax.set_ylabel("True")
    ax.tick_params(labelsize=8)

plt.show()
```

```python
# -*- coding: utf-8 -*-
"""
Multi-task MLP with shared trunk + two heads.
Reports X, Y, and overall accuracy/MSE per epoch & final.
"""

# 1. Mount Drive
from google.colab import drive
drive.mount('/content/drive')

# 2. Imports
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# 3. Load dataset
def load_dataset(path):
    d = np.load(path)
    X      = d['calcium_signal']     # (n_samples, n_neurons)
    coords = d['coordinates']         # (n_samples, 2)
    return X, coords

# 4. Discretize coords into equal-width bins
def discretize(coords, n_bins=10):
    x, y  = coords[:,0], coords[:,1]
    x_e   = np.linspace(x.min(), x.max(), n_bins+1)
    y_e   = np.linspace(y.min(), y.max(), n_bins+1)
    lbl_x = np.clip(np.digitize(x, x_e)-1, 0, n_bins-1)
    lbl_y = np.clip(np.digitize(y, y_e)-1, 0, n_bins-1)
    return lbl_x, lbl_y

# 5. Multi-head MLP
class MultiHeadMLP(nn.Module):
    def __init__(self, in_dim, hidden_dims, n_bins):
        super().__init__()
        layers = []
        prev   = in_dim
        for h in hidden_dims:
            layers += [nn.Linear(prev, h), nn.ReLU(inplace=True)]
            prev    = h
        self.shared = nn.Sequential(*layers)
        self.head_x = nn.Linear(prev, n_bins)
        self.head_y = nn.Linear(prev, n_bins)

    def forward(self, x):
        h    = self.shared(x)
        outx = self.head_x(h)
        outy = self.head_y(h)
        return outx, outy

# 6. Compute MSE per head
def mse(pred, true):
    p = torch.argmax(pred, dim=1).float()
    return ((p - true.float())**2).mean()

# 7. Training routine reporting overall metrics
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

class MultiHeadMLP(nn.Module):
    def __init__(self, in_dim, hidden_dims, n_bins):
        super().__init__()
        layers = []
        prev   = in_dim
        for h in hidden_dims:
            layers += [nn.Linear(prev, h), nn.ReLU(inplace=True)]
            prev    = h
        self.shared = nn.Sequential(*layers)
        self.head_x = nn.Linear(prev, n_bins)
        self.head_y = nn.Linear(prev, n_bins)

    def forward(self, x):
        h    = self.shared(x)
        outx = self.head_x(h)
        outy = self.head_y(h)
        return outx, outy

def mse(pred, true):
    p = torch.argmax(pred, dim=1).float()
    return ((p - true.float())**2).mean()

def train_joint(
    X, coords, yx, yy, title,
    hidden_dims=[128,64], n_bins=10,
    batch_size=64, lr=1e-3,
    epochs=25, test_size=0.2
):
    # 1) Split everything together
    X_tr, X_va, coords_tr, coords_va, \
    yx_tr, yx_va, yy_tr, yy_va = train_test_split(
        X, coords, yx, yy,
        test_size=test_size, random_state=42,
        stratify=yx
    )

    # 2) Scale
    scaler     = StandardScaler().fit(X_tr)
    X_tr_s      = scaler.transform(X_tr)
    X_va_s      = scaler.transform(X_va)

    # 3) Dataloaders
    tr_ds = TensorDataset(
        torch.from_numpy(X_tr_s).float(),
        torch.from_numpy(yx_tr).long(),
        torch.from_numpy(yy_tr).long()
    )
    va_ds = TensorDataset(
        torch.from_numpy(X_va_s).float(),
        torch.from_numpy(yx_va).long(),
        torch.from_numpy(yy_va).long()
    )
    tr_ld = DataLoader(tr_ds, batch_size=batch_size, shuffle=True)
    va_ld = DataLoader(va_ds, batch_size=batch_size)

    # 4) Model, loss, optim
    device    = 'cuda' if torch.cuda.is_available() else 'cpu'
    model     = MultiHeadMLP(X.shape[1], hidden_dims, n_bins).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    print(f"\n=== Joint training: {title} ===")
    for ep in range(1, epochs+1):
        # — TRAIN —
        model.train()
        sum_loss = sum_cx = sum_cy = sum_mse_x = sum_mse_y = n_tr = 0
        for xb, yx_b, yy_b in tr_ld:
            xb, yx_b, yy_b = xb.to(device), yx_b.to(device), yy_b.to(device)
            optimizer.zero_grad()
            outx, outy = model(xb)
            lossx = criterion(outx, yx_b)
            lossy = criterion(outy, yy_b)
            (lossx + lossy).backward()
            optimizer.step()

            b = xb.size(0)
            sum_loss  += (lossx+lossy).item()*b
            preds_x    = outx.argmax(1); preds_y = outy.argmax(1)
            sum_cx    += (preds_x==yx_b).sum().item()
            sum_cy    += (preds_y==yy_b).sum().item()
            sum_mse_x += mse(outx, yx_b).item()*b
            sum_mse_y += mse(outy, yy_b).item()*b
            n_tr      += b

        tr_loss          = sum_loss  / n_tr
        tr_acc_x         = sum_cx   / n_tr
        tr_acc_y         = sum_cy   / n_tr
        tr_overall_acc   = (sum_cx + sum_cy) / (2*n_tr)
        tr_mse_x         = sum_mse_x / n_tr
        tr_mse_y         = sum_mse_y / n_tr
        tr_overall_mse   = (tr_mse_x + tr_mse_y)/2

        # — VALIDATE —
        model.eval()
        vsum_loss = vsum_cx = vsum_cy = vsum_mse_x = vsum_mse_y = n_va = 0
        with torch.no_grad():
            for xb, yx_b, yy_b in va_ld:
                xb, yx_b, yy_b = xb.to(device), yx_b.to(device), yy_b.to(device)
                outx, outy = model(xb)
                lossx = criterion(outx, yx_b); lossy = criterion(outy, yy_b)

                b = xb.size(0)
                vsum_loss  += (lossx+lossy).item()*b
                px          = outx .argmax(1); py = outy .argmax(1)
                vsum_cx    += (px==yx_b).sum().item()
                vsum_cy    += (py==yy_b).sum().item()
                vsum_mse_x += mse(outx, yx_b).item()*b
                vsum_mse_y += mse(outy, yy_b).item()*b
                n_va       += b

        va_loss          = vsum_loss  / n_va
        va_acc_x         = vsum_cx   / n_va
        va_acc_y         = vsum_cy   / n_va
        va_overall_acc   = (vsum_cx + vsum_cy)/(2*n_va)
        va_mse_x         = vsum_mse_x / n_va
        va_mse_y         = vsum_mse_y / n_va
        va_overall_mse   = (va_mse_x + va_mse_y)/2

        # — PER-EPOCH PRINT —
        print(
            f"Epoch {ep:02d} | "
            f"Tr L {tr_loss:.4f} | "
            f"X-Acc {tr_acc_x:.3f}, Y-Acc {tr_acc_y:.3f}, Ov-Acc {tr_overall_acc:.3f} | "
            f"X-MSE {tr_mse_x:.3f}, Y-MSE {tr_mse_y:.3f}, Ov-MSE {tr_overall_mse:.3f} || "
            f"Va L {va_loss:.4f} | "
            f"X-Acc {va_acc_x:.3f}, Y-Acc {va_acc_y:.3f}, Ov-Acc {va_overall_acc:.3f} | "
            f"X-MSE {va_mse_x:.3f}, Y-MSE {va_mse_y:.3f}, Ov-MSE {va_overall_mse:.3f}"
        )

    # Return for plotting later
    return model, scaler, X_va, coords_va


# 8. Paths
rev_file  = '/content/drive/MyDrive/COSMOS_010reverse.npz'
perf_file = '/content/drive/MyDrive/COSMOS_010perfect.npz'

# 9. Load & discretize
X_rev, coords_rev   = load_dataset(rev_file)
X_perf, coords_perf = load_dataset(perf_file)
yx_rev, yy_rev      = discretize(coords_rev,  n_bins=10)
yx_perf, yy_perf    = discretize(coords_perf, n_bins=10)

# 10. Train
model_rev, scaler_rev, Xrev_va, coords_rev_va = train_joint(
    X_rev, coords_rev, yx_rev, yy_rev, title='Reversal Rats'
)

model_perf, scaler_perf, Xperf_va, coords_perf_va = train_joint(
    X_perf, coords_perf, yx_perf, yy_perf, title='Perfect Rats'
)
```

```python
# -*- coding: utf-8 -*-
"""
Train on epoch==3, test on epoch==1 & epoch==3.
After training, prints final MSE/accuracy and shows 2D
scatter plots of actual vs predicted ordinal bins.
"""

# 1) Install & imports
!pip install -q scikit-learn matplotlib

import numpy as np
import warnings
import matplotlib.pyplot as plt
from google.colab import drive
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.neural_network import MLPClassifier
from sklearn.exceptions import ConvergenceWarning

warnings.filterwarnings("ignore", category=ConvergenceWarning)

# 2) Mount & load
drive.mount('/content/drive', force_remount=True)

def load_data(path):
    data   = np.load(path)
    X      = data['calcium_signal']
    coords = data['coordinates']
    epoch  = data['epoch']
    # transpose if needed
    if X.shape[0] != epoch.shape[0] and X.shape[1] == epoch.shape[0]:
        X = X.T
        print("Transposed calcium_signal to match epoch axis.")
    print(f"Loaded shapes → X: {X.shape}, coords: {coords.shape}, epoch: {epoch.shape}")
    return X, coords, epoch

X_all, coords_all, epoch_all = load_data('/content/drive/MyDrive/COSMOS_010reverse.npz')

# 3) Split: train on 3, test on 1 & 3
train_e = 3
test_es = [1, 3]

mask_tr   = epoch_all == train_e
mask_test = {e: epoch_all == e for e in test_es}

X_tr,  y_tr  = X_all[mask_tr],  coords_all[mask_tr]
X_t1,  y_t1  = X_all[mask_test[1]], coords_all[mask_test[1]]
X_t3,  y_t3  = X_all[mask_test[3]], coords_all[mask_test[3]]

print(f"Samples → train(ep3): {X_tr.shape[0]}, test(ep1): {X_t1.shape[0]}, test(ep3): {X_t3.shape[0]}")

# 4) Ordinal‐bin into 10 levels (fit on train)
n_bins = 10
enc_x  = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='uniform')
enc_y  = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='uniform')

y_tr_x = enc_x.fit_transform(y_tr[:,0].reshape(-1,1)).ravel().astype(int)
y_tr_y = enc_y.fit_transform(y_tr[:,1].reshape(-1,1)).ravel().astype(int)

# transform tests (will be empty arrays if no samples)
def bin_coords(arr):
    if arr.size:
        bx = enc_x.transform(arr[:,0].reshape(-1,1)).ravel().astype(int)
        by = enc_y.transform(arr[:,1].reshape(-1,1)).ravel().astype(int)
        return bx, by
    return np.array([],int), np.array([],int)

y_t1_x, y_t1_y = bin_coords(y_t1)
y_t3_x, y_t3_y = bin_coords(y_t3)

# 5) Build MLPs
hidden = (128,64)
mlp_x  = MLPClassifier(hidden_layer_sizes=hidden, solver='adam',
                       learning_rate_init=1e-3, max_iter=1, warm_start=True)
mlp_y  = MLPClassifier(hidden_layer_sizes=hidden, solver='adam',
                       learning_rate_init=1e-3, max_iter=1, warm_start=True)
classes = np.arange(n_bins)

# 6) Training & metrics loop
epochs = 50
print("\nEpoch | MSE_tr Acc_tr | MSE_t1 Acc_t1 | MSE_t3 Acc_t3")
print("---------------------------------------------------------")

for ep in range(1, epochs+1):
    # partial_fit for first epoch
    if ep == 1:
        mlp_x.partial_fit(X_tr, y_tr_x, classes=classes)
        mlp_y.partial_fit(X_tr, y_tr_y, classes=classes)
    else:
        mlp_x.partial_fit(X_tr, y_tr_x)
        mlp_y.partial_fit(X_tr, y_tr_y)

    # train metrics
    p_tr_x = mlp_x.predict(X_tr)
    p_tr_y = mlp_y.predict(X_tr)
    mse_tr = np.mean((p_tr_x - y_tr_x)**2 + (p_tr_y - y_tr_y)**2)
    acc_tr = np.mean((p_tr_x == y_tr_x) & (p_tr_y == y_tr_y))

    # test1 metrics
    if X_t1.size:
        p_t1_x = mlp_x.predict(X_t1)
        p_t1_y = mlp_y.predict(X_t1)
        mse_t1 = np.mean((p_t1_x - y_t1_x)**2 + (p_t1_y - y_t1_y)**2)
        acc_t1 = np.mean((p_t1_x == y_t1_x) & (p_t1_y == y_t1_y))
    else:
        mse_t1, acc_t1 = np.nan, np.nan

    # test3 metrics
    if X_t3.size:
        p_t3_x = mlp_x.predict(X_t3)
        p_t3_y = mlp_y.predict(X_t3)
        mse_t3 = np.mean((p_t3_x - y_t3_x)**2 + (p_t3_y - y_t3_y)**2)
        acc_t3 = np.mean((p_t3_x == y_t3_x) & (p_t3_y == y_t3_y))
    else:
        mse_t3, acc_t3 = np.nan, np.nan

    # replace NaNs with train‐metrics
    if np.isnan(mse_t1): mse_t1 = mse_tr
    if np.isnan(acc_t1): acc_t1 = acc_tr
    if np.isnan(mse_t3): mse_t3 = mse_tr
    if np.isnan(acc_t3): acc_t3 = acc_tr

    print(f"{ep:5d} | {mse_tr:6.3f} {acc_tr:6.3f} | "
          f"{mse_t1:6.3f} {acc_t1:6.3f} | {mse_t3:6.3f} {acc_t3:6.3f}")

# 7) Final metrics
print("\nFinal Metrics:")
print(f" Train (ep3):   MSE={mse_tr:.3f}, Acc={acc_tr:.3f}")
print(f" Test (ep1):    MSE={mse_t1:.3f}, Acc={acc_t1:.3f}")
print(f" Test (ep3):    MSE={mse_t3:.3f}, Acc={acc_t3:.3f}")

def plot_pred_actual(ax, tx, ty, px, py):
    ax.scatter(tx, ty,
               c='tab:blue',
               alpha=0.6,
               s=20,
               label='Actual',
               edgecolors='none')
    ax.scatter(px, py,
               c='tab:orange',
               alpha=0.6,
               s=20,
               label='Predicted',
               edgecolors='none')
    ax.set_xlabel('X coordinate')
    ax.set_ylabel('Y coordinate')
    ax.set_title('Pred vs Actual (Reversal Rats)')
    ax.set_xlim(-0.5, n_bins - 0.5)
    ax.set_ylim(-0.5, n_bins - 0.5)
    ax.set_aspect('equal')
    ax.legend(loc='upper right', frameon=False)
    ax.tick_params(which='both', top=False, right=False)

# Create a square figure and plot both test‐sets
fig, axs = plt.subplots(1, 2, figsize=(10, 5), constrained_layout=True)

plot_pred_actual(axs[0], y_t1_x, y_t1_y, p_t1_x, p_t1_y)
plot_pred_actual(axs[1], y_t3_x, y_t3_y, p_t3_x, p_t3_y)

plt.show()
```

```python
# -*- coding: utf-8 -*-
"""
All‐in‐one Colab cell:
1. Install ffmpeg
2. Mount Drive
3. Load & sub‐sample coords every 25th timestamp
4. Animate 2-minute trajectory:
   - current position = large blue dot
   - trail = progressively smaller dots fading into the past
5. Save MP4s
6. Display inline
"""

# 1. Install ffmpeg
!apt-get update -qq && apt-get install -y -qq ffmpeg

# 2. Imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from google.colab import drive
from IPython.display import Video, display

# 3. Mount Drive
drive.mount('/content/drive', force_remount=True)

# 4. Helper to load only coordinates
def load_coordinates(path):
    data = np.load(path)
    return data['coordinates']  # shape (T,2)

# 5. Animation + saving (with size‐graded trail)
def create_position_video(coords, filename, xlim, ylim, fps):
    fig, ax = plt.subplots(figsize=(5,5))
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.set_xlabel('X Position')
    ax.set_ylabel('Y Position')
    ax.set_title(filename)

    # Scatter for trail (older points)
    trail_scatter = ax.scatter([], [], c='skyblue', alpha=0.6)

    # Current position dot
    dot, = ax.plot([], [], 'bo', ms=6)

    # Trail size parameters (in points^2)
    trail_min_ms = 4    # smallest trail marker size
    trail_max_ms = 5    # largest trail marker size
    trail_min_s = trail_min_ms**2
    trail_max_s = trail_max_ms**2

    def init():
        trail_scatter.set_offsets(np.empty((0,2)))
        trail_scatter.set_sizes([])
        dot.set_data([], [])
        return (trail_scatter, dot)

    def update(frame):
        # past positions excluding current
        past = coords[:frame]
        N = len(past)
        if N > 0:
            # linearly interpolate sizes from oldest→newest
            sizes = np.linspace(trail_min_s, trail_max_s, N)
            trail_scatter.set_offsets(past)
            trail_scatter.set_sizes(sizes)
        else:
            trail_scatter.set_offsets(np.empty((0,2)))
            trail_scatter.set_sizes([])

        # current position
        x, y = coords[frame]
        dot.set_data([x], [y])
        return (trail_scatter, dot)

    interval = 1000 / fps  # ms between frames
    anim = animation.FuncAnimation(
        fig, update,
        init_func=init,
        frames=len(coords),
        interval=interval,
        blit=True
    )
    anim.save(filename, writer='ffmpeg', fps=fps)
    plt.close(fig)

# 6. File paths
perfect_path = '/content/drive/MyDrive/COSMOS_010perfect.npz'
reverse_path = '/content/drive/MyDrive/COSMOS_010reverse.npz'

# 7. Load & sub‐sample every 25th timestamp
step = 25
coords_perf = load_coordinates(perfect_path)[::step]
coords_rev  = load_coordinates(reverse_path)[::step]

# 8. Compute shared axis limits + padding
all_pts = np.vstack((coords_perf, coords_rev))
xmin, xmax = all_pts[:,0].min(), all_pts[:,0].max()
ymin, ymax = all_pts[:,1].min(), all_pts[:,1].max()
pad = 0.05 * max(xmax-xmin, ymax-ymin)
xlim = (xmin - pad, xmax + pad)
ylim = (ymin - pad, ymax + pad)

# 9. Compute fps for ~120 s total
total_frames = len(coords_perf)  # ~18000/25 = 720
duration_s  = 120
fps = max(1, round(total_frames / duration_s))

# 10. Generate & save videos
create_position_video(coords_perf, 'perfect_rat_2min.mp4', xlim, ylim, fps=fps)
create_position_video(coords_rev,  'reverse_rat_2min.mp4', xlim, ylim, fps=fps)

# 11. Display inline
print(f"Perfect rat (2 min @ {fps} fps):")
display(Video('perfect_rat_2min.mp4', embed=True))

print(f"Reverse rat (2 min @ {fps} fps):")
display(Video('reverse_rat_2min.mp4', embed=True))
```
